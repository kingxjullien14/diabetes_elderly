## 1. Importing Libraries

```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Libraries imported successfully!")
```

## 2. Loading the Data
Loading both 2023 and 2024 BRFSS datasets

```python
# Load 2023 data
data_2023 = pd.read_csv('2023_BRFSS_CLEANED.csv', encoding='utf-8')
print(f"2023 Data Shape: {data_2023.shape}")
print(f"Columns: {data_2023.columns.tolist()}")
print("\n2023 Data Sample:")
display(data_2023.head())
```

```python
# Load 2024 data
data_2024 = pd.read_csv('2024_BRFSS_CLEANED.csv', encoding='utf-8')
print(f"2024 Data Shape: {data_2024.shape}")
print(f"Columns: {data_2024.columns.tolist()}")
print("\n2024 Data Sample:")
display(data_2024.head())
```

```python
# Combine both datasets
data = pd.concat([data_2023, data_2024], ignore_index=True)
print(f"\nCombined Data Shape: {data.shape}")
print(f"Total records: {len(data):,}")
print(f"2023 records: {len(data_2023):,}")
print(f"2024 records: {len(data_2024):,}")
```

## 3. Initial Data Exploration

```python
# Data info
print("Data Information:")
data.info()
```

```python
# Statistical summary
print("Statistical Summary:")
display(data.describe())
```

```python
# Check target variable distribution
print("Diabetes Status Distribution:")
print(data['DIABETES_STATUS'].value_counts())
print("\nPercentage:")
print(data['DIABETES_STATUS'].value_counts(normalize=True) * 100)
```

## 4. Handling Missing Values

```python
# Check for missing values
print("Missing Values Summary:")
missing_summary = pd.DataFrame({
    'Column': data.columns,
    'Missing_Count': data.isnull().sum().values,
    'Missing_Percentage': (data.isnull().sum().values / len(data) * 100).round(2)
})
missing_summary = missing_summary[missing_summary['Missing_Count'] > 0].sort_values('Missing_Percentage', ascending=False)
display(missing_summary)
```

```python
# Visualize missing data
plt.figure(figsize=(12, 6))
missing_pct = (data.isnull().sum() / len(data) * 100).sort_values(ascending=False)
missing_pct = missing_pct[missing_pct > 0]
if len(missing_pct) > 0:
    missing_pct.plot(kind='bar', color='coral')
    plt.title('Missing Values Percentage by Column', fontsize=14, fontweight='bold')
    plt.ylabel('Percentage (%)')
    plt.xlabel('Columns')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("No missing values found!")
```

```python
# Strategy for handling missing values
# For columns with high missing percentage (>30%), consider dropping
# For columns with moderate missing (10-30%), use median/mode imputation
# For columns with low missing (<10%), use forward/backward fill or mode

# Let's first examine which columns have what percentage
missing_high = missing_summary[missing_summary['Missing_Percentage'] > 30]
missing_moderate = missing_summary[(missing_summary['Missing_Percentage'] > 10) & (missing_summary['Missing_Percentage'] <= 30)]
missing_low = missing_summary[missing_summary['Missing_Percentage'] <= 10]

print(f"Columns with high missing (>30%): {list(missing_high['Column'])}")
print(f"Columns with moderate missing (10-30%): {list(missing_moderate['Column'])}")
print(f"Columns with low missing (<10%): {list(missing_low['Column'])}")
```

```python
# Create a copy for processing
data_clean = data.copy()

# Drop rows where target variable is missing
if data_clean['DIABETES_STATUS'].isnull().sum() > 0:
    print(f"Dropping {data_clean['DIABETES_STATUS'].isnull().sum()} rows with missing target variable")
    data_clean = data_clean[data_clean['DIABETES_STATUS'].notna()]

# For numerical columns with missing values, use median imputation
numerical_cols = data_clean.select_dtypes(include=[np.number]).columns.tolist()
numerical_cols.remove('DIABETES_STATUS')  # Don't impute target

for col in numerical_cols:
    if data_clean[col].isnull().sum() > 0:
        median_value = data_clean[col].median()
        print(f"Imputing {col} with median: {median_value}")
        data_clean[col].fillna(median_value, inplace=True)

# For categorical columns, use mode imputation
categorical_cols = data_clean.select_dtypes(include=['object']).columns.tolist()
for col in categorical_cols:
    if data_clean[col].isnull().sum() > 0:
        mode_value = data_clean[col].mode()[0]
        print(f"Imputing {col} with mode: {mode_value}")
        data_clean[col].fillna(mode_value, inplace=True)

print(f"\nMissing values after imputation: {data_clean.isnull().sum().sum()}")
```

## 5. Checking and Removing Duplicates

```python
# Check for duplicates
duplicates = data_clean.duplicated().sum()
print(f"Number of duplicate rows: {duplicates:,}")

if duplicates > 0:
    print(f"Percentage of duplicates: {(duplicates/len(data_clean)*100):.2f}%")
    data_clean = data_clean.drop_duplicates()
    print(f"Shape after removing duplicates: {data_clean.shape}")
else:
    print("No duplicates found!")
```

## 6. Data Type Conversion and Encoding

```python
# Check current data types
print("Current Data Types:")
print(data_clean.dtypes)
```

```python
# Convert all numeric columns to appropriate integer type
# Most BRFSS variables are categorical encoded as integers
for col in data_clean.columns:
    if col not in ['WGHT (lbs)', 'HGHT (ft)', 'BMI']:  # Keep these as float
        if data_clean[col].dtype in ['float64', 'float32']:
            data_clean[col] = data_clean[col].astype(int)

print("\nData types after conversion:")
print(data_clean.dtypes)
```

## 7. Outlier Detection and Removal

```python
# Visualize potential outliers in key continuous variables
continuous_vars = ['WGHT (lbs)', 'HGHT (ft)', 'BMI', 'PHYS_HLTH_DAYS', 'MENT_HLTH_DAYS', 'POOR_HLTH_DAYS']
existing_continuous = [col for col in continuous_vars if col in data_clean.columns]

if existing_continuous:
    fig, axes = plt.subplots(2, 3, figsize=(15, 8))
    axes = axes.flatten()
    
    for idx, col in enumerate(existing_continuous):
        if idx < len(axes):
            axes[idx].boxplot(data_clean[col].dropna())
            axes[idx].set_title(f'Boxplot of {col}')
            axes[idx].set_ylabel(col)
    
    # Hide unused subplots
    for idx in range(len(existing_continuous), len(axes)):
        axes[idx].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
# Use Isolation Forest for outlier detection
from sklearn.ensemble import IsolationForest

# Select only numeric columns for outlier detection
numeric_cols_for_outliers = data_clean.select_dtypes(include=[np.number]).columns.tolist()

print(f"Before outlier removal: {data_clean.shape}")

# Fit Isolation Forest
model = IsolationForest(contamination=0.1, random_state=42)  # Expect ~10% outliers
data_clean['anomaly'] = model.fit_predict(data_clean[numeric_cols_for_outliers])

# Count outliers
outliers = data_clean[data_clean['anomaly'] == -1]
print(f"Number of outliers detected: {len(outliers):,}")
print(f"Percentage of outliers: {(len(outliers)/len(data_clean)*100):.2f}%")

# Remove outliers
data_clean = data_clean[data_clean['anomaly'] == 1]
data_clean.drop(columns=['anomaly'], inplace=True)

print(f"After outlier removal: {data_clean.shape}")
```

## 8. Feature Engineering

```python
# Create BMI categories if needed
if 'BMI' in data_clean.columns:
    data_clean['BMI_CATEGORY'] = pd.cut(data_clean['BMI'], 
                                         bins=[0, 18.5, 25, 30, 100],
                                         labels=[0, 1, 2, 3])  # 0=Underweight, 1=Normal, 2=Overweight, 3=Obese
    print("BMI Categories Created:")
    print(data_clean['BMI_CATEGORY'].value_counts().sort_index())

# Create age groups if needed
if 'AGE' in data_clean.columns:
    data_clean['AGE_GROUP'] = pd.cut(data_clean['AGE'], 
                                      bins=[0, 2, 4, 6, 13],
                                      labels=[0, 1, 2, 3])  # Based on BRFSS age categories
    print("\nAge Groups Created:")
    print(data_clean['AGE_GROUP'].value_counts().sort_index())
```

## 9. Creating Binary Target Variable

```python
# Create binary diabetes variable (0 = No Diabetes, 1 = Diabetes)
# BRFSS DIABETES_STATUS: 0=No, 1=Yes, 2=Borderline/Prediabetes, 3=Yes during pregnancy

print("Original DIABETES_STATUS distribution:")
print(data_clean['DIABETES_STATUS'].value_counts().sort_index())

# Create binary variable: 0 = No diabetes, 1 = Has diabetes (including prediabetes)
# You can adjust this based on your research question
data_clean['DIABETES_BINARY'] = data_clean['DIABETES_STATUS'].apply(
    lambda x: 0 if x == 0 else 1
)

print("\nBinary DIABETES_BINARY distribution:")
print(data_clean['DIABETES_BINARY'].value_counts())
print("\nPercentage:")
print(data_clean['DIABETES_BINARY'].value_counts(normalize=True) * 100)
```

## 10. Save Preprocessed Data

```python
# Save the cleaned data
output_file = 'BRFSS_preprocessed.csv'
data_clean.to_csv(output_file, index=False)
print(f"Preprocessed data saved to: {output_file}")
print(f"Final shape: {data_clean.shape}")
print(f"\nColumns: {data_clean.columns.tolist()}")
```
